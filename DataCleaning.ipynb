```python
import pandas as pd

# Load the gapminder_full.csv file
data = pd.read_csv('data/gapminder_full.csv')

# Display the first few rows of the dataframe
data.head()
```

```python
import pycountry

# Function to get the alpha-3 code for a country
def get_alpha_3(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_3
    except LookupError:
        return None

# Apply the function to the 'country' column to create the 'iso_alpha' column
data['iso_alpha'] = data['country'].apply(get_alpha_3)

# Display the first few rows of the dataframe to verify the new column
data.head()
```

```python
# Filter rows with missing data in iso_alpha column
# Don't run this cell if you want to keep the rest of the values?
data = data[data['iso_alpha'].isna()]
data.head()
```

```python
data.loc[data['country'] == 'Congo, Dem. Rep.', 'iso_alpha'] = 'COD'
data.loc[data['country'] == 'Congo, Rep.', 'iso_alpha'] = 'COG'
data.loc[data['country'] == "Cote d'Ivoire", 'iso_alpha'] = 'CIV'
data.loc[data['country'] == "Hong Kong, China", 'iso_alpha'] = 'HKG'
data.loc[data['country'] == "Korea, Dem. Rep.", 'iso_alpha'] = 'PRK'
data.loc[data['country'] == "Korea, Rep.", 'iso_alpha'] = 'KOR'
data.loc[data['country'] == "Reunion", 'iso_alpha'] = 'REU'
data.loc[data['country'] == "Swaziland", 'iso_alpha'] = 'SWZ'
data.loc[data['country'] == "Turkey", 'iso_alpha'] = 'TUR'
data.loc[data['country'] == "West Bank and Gaza", 'iso_alpha'] = 'PSE'
data.loc[data['country'] == "Yemen, Rep.", 'iso_alpha'] = 'YEM'
data.head()
```
